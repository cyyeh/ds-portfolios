# Baseline Model

## Import Libraries

```python
# Data and Stats packages
import numpy as np
import pandas as pd

# Visualization packages
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

## Data Wrangling

### Loading Data

```python
train_data_path = 'data/train.csv'
test_data_path = 'data/test.csv'

train_data = pd.read_csv(train_data_path, index_col=0)
test_data = pd.read_csv(test_data_path, index_col=0)
```

```python
train_data
```

```python
train_data.info()
```

### Imputation: drop columns that have NaN

```python
new_train_data = train_data.dropna(axis=1)
```

```python
new_train_data.info()
```

```python
y_train = new_train_data.iloc[:, -1]
```

```python
X_train = pd.get_dummies(new_train_data.iloc[:, :-1])
```

```python
linear_reg = LinearRegression()
scores = cross_val_score(linear_reg, X_train, y_train, cv=10)
```

```python
np.mean(scores)
```

```python
linear_reg.fit(X_train, y_train)
```

```python
test_data
```

```python
new_test_data = test_data.dropna(axis=1)
```

```python
new_test_data.info()
```

```python
X_test = pd.get_dummies(new_test_data)
```

```python
linear_reg.predict(X_test)
```
